```python
#importing libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusting the configuration of the plot
```

```python
#Read in the DataSet

df = pd.read_csv(r'C:\Users\welcome\Downloads\movies.csv',encoding='latin-1')
```

```python
#Looking at the data
df.head()
```

```python
#Checking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
```

```python
#Data type for all columns 

df.dtypes
```

```python
#Changing Data type from float to integer

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df['CorrectYear'] = df['released'].astype(str).str[:4]
df.head()
```

```python
#pd.set_option('display.max_rows',None)
```

```python
df = df.sort_values(by=['gross'],inplace=False,ascending = False)
df.head()
```

```python
#Checking any duplicates to be dropped

df['company'].drop_duplicates().sort_values(ascending=False)
```

Finding corelation in Data

```python
 #Sctter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('BUDGET vs GROSS Earnings')

plt.xlabel('GROSS EARNINGS')

plt.ylabel('FILM BUDGET')

plt.show()
```

```python
#Plot Budget vs Gross using Seaborn

sns.regplot(x='budget',y='gross',data=df,scatter_kws ={"color":'red'},line_kws = {'color':"blue"})
```

Looking at correlation

```python
df.corr() #pearson--Default ,kendall,spearman
```

#High Correlation between BUDGET AND GROSS

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix,annot=True)

plt.title('Corelation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#Look At Company
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix,annot=True)

plt.title('Corelation Matrix_numerized')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
#High Corrlation 
High_corr = sorted_pairs[(sorted_pairs)>0.5]
print(High_corr)
```
